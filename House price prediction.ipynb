```python
# treating the outliers
def outliers(df,variable):
    IQR=df[variable].quantile(0.75)-df[variable].quantile(0.25)
    LB=df[variable].quantile(0.25)-(IQR*1.5)
    UB=df[variable].quantile(0.75)+(IQR*1.5)
    df.loc[df[variable]>=UB,variable]=UB
```

```python
outliers(df_num,'MSSubClass')
outliers(df_num,'LotFrontage')
outliers(df_num,'LotArea')
outliers(df_num,'BsmtFinSF1')
outliers(df_num,'BsmtUnfSF')
outliers(df_num,'TotalBsmtSF')
outliers(df_num,'1stFlrSF')
outliers(df_num,'GrLivArea')
outliers(df_num,'GarageArea')
outliers(df_num,'OpenPorchSF')
outliers(df_num,'TotRmsAbvGrd')    
```
